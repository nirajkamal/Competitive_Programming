## Question-1
Given an array A and an integer B. A pair(i, j) in the array is a good pair if i != j and (A[i] + A[j] == B). Check if any good pair exist or not.

Problem Constraints
1 <= A.size() <= 104

1 <= A[i] <= 109

1 <= B <= 109
<br><br>


Input Format <br>
First argument is an integer array A.

Second argument is an integer B.
<br><br>


Output Format <br>
Return 1 if good pair exist otherwise return 0.

<br><br>

Example Input <br>
Input 1:

A = [1,2,3,4]
B = 7
Input 2:

A = [1,2,4]
B = 4
Input 3:

A = [1,2,2]
B = 4

<br><br>
Example Output<br>
Output 1:
1

Output 2:
0

Output 3:
1

<br><br>
Example Explanation<br>
Explanation 1:

 (i,j) = (3,4)
Explanation 2:

No pair has sum equal to 4.
Explanation 3:

 (i,j) = (2,3)



```python
#O(n) solution

class Solution:
    # Param A: list of integers
    # Param B: integer
    # return an integer

    def solve(self, A,B):
        elementSum = {}
        for i in A:
            if i in elementSum:
                elementSum[i] +=1
            else:
                elementSum[i] = 1
        for i in A:
            if B-i in elementSum:
                if B-i == i and elementSum[i]>1:
                    return (1)
                elif B-i != i:
                    return(1)
        return(0)
```

```python
# O(logn) solution.

class Solution:
    # @param A : list of integers
    # @param B : integer
    # @return an integer
    def solve(self, A, B):
        A.sort()
        for i in range(len(A)):
            for j in range(i+1, len(A)):
                if j == i+1:
                    if A[i]+A[j]>B:
                        return(0)
                if A[i]+A[j]==B:
                    return(1)

        return(0)
```

```python
def rotateArray(A,s,e):
    N = e-s+1
    for i in range(N//2):
        (A[i+s],A[e-i]) = (A[e-i],A[i+s])
    return(A)


def main():
    # YOUR CODE GOES HERE
    # Please take input and print output to standard input/output (stdin/stdout)
    # E.g. 'input()/raw_input()' for input & 'print' for output
    A = input()
    B = input()
    B = int(B)
    A = A.split(' ')
    A.pop(0)
    A = [int(x) for x in A if x!= '']
    N = len(A)
    #rotate entire arrray
    A = rotateArray(A,0,N-1)
    #find if B<N, and B>N:
    B = B%N
    #rotate first B half of the array
    A = rotateArray(A,0,B-1)
    #rotate second B half of the array
    A = rotateArray(A,B,N-1)
    A= [str(x) for x in A]
    A= ' '.join(A)
    A = A+' '
    print(A)
    return 0

if __name__ == '__main__':
    main()
```
